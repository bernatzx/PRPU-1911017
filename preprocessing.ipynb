```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import joblib
```

```python
DATA_PATH = 'data/data_mentah.xlsx'
TRAIN_PATH = 'data/data_train.csv'
TEST_PATH = 'data/data_test.csv'
SCALER_PATH = 'data/scaler.pkl'
TARGET_ENCODER_PATH = 'data/target_encoder.pkl'
```

```python
df = pd.read_excel(DATA_PATH)
# Hapus kolom ID karena tidak diperlukan
if 'ID' in df.columns:
    df.drop(columns=['ID'], inplace=True)

# Menentukan kolom target dalam hal ini kolom kelayakan
target_column = 'KELAYAKAN'
# Pisahkan target
y = df[target_column]
x = df.drop(columns=[target_column])
```

ENCODING FITUR KATEGORIKAL

```python
# Encode target (Layak = 1, Tidak Layak = 0)
target_encoder = LabelEncoder()
y_encoded = target_encoder.fit_transform(y)
joblib.dump(target_encoder, TARGET_ENCODER_PATH)
```

```python
# Mapping hasil encoding untuk melihat hasilnya
df['KELAYAKAN_ENCODED'] = target_encoder.fit_transform(df['KELAYAKAN'])

# Tampilkan hasil
display(df[['KELAYAKAN', 'KELAYAKAN_ENCODED']].drop_duplicates())  # untuk lihat mapping
df.tail() # untuk lihat hasil pada data
```

```python
# One-hot encoding kolom kategorikal
# Menentukan kolom numerik dan kategorikal
numeric_features = ['PENGHASILAN', 'TANGGUNGAN', 'USIA', 'BESAR PINJAMAN', 'TENOR', 'DP PRESENTASE'] # numerik
categorical_features = ['PEKERJAAN', 'JAMINAN', 'JENIS KREDIT', 'STATUS KEPEMILIKAN RUMAH', 'RIWAYAT KREDIT'] # kategorikal

 # One-hot encoding untuk kolom kategorikal
X_encoded = pd.get_dummies(x, columns=categorical_features)
```

```python
# Menampilkan hasil one-hot encoding pada kolom kategorikal
X_encoded.head()
```

NORMALISASI KOLOM NUMERIK

```python
scaler = StandardScaler()
X_encoded[numeric_features] = scaler.fit_transform(X_encoded[numeric_features])
joblib.dump(scaler, SCALER_PATH)
```

```python
X_encoded.head()
```

PEMBAGIAN DATA

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded, test_size=0.2, random_state=42
)

# Tampilkan ukuran data
print("Ukuran data latih:", X_train.shape)
print("Ukuran data uji:", X_test.shape)
```

MENYIMPAN DATA LATIH DAN UJI

```python
# Data Latih
train_df = pd.DataFrame(X_train)
train_df['target'] = y_train
train_df.to_csv(TRAIN_PATH, index=False)

# Data Uji
test_df = pd.DataFrame(X_test)
test_df['target'] = y_test
test_df.to_csv(TEST_PATH, index=False)
```
